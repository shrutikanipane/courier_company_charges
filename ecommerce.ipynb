```python
#slab for x and courier company
def total_slab_weight(weight):
    if(weight>0 and weight<=0.5):
        return 0.5
    elif(weight>0.5 and weight<=1.0):
        return 1
    elif(weight>1 and weight<=1.5):
        return 1.5
    elif(weight>1.5 and weight<=2.0):
        return 2
    elif(weight>2.0 and weight<=2.5):
        return 2.5
    elif(weight>2.0 and weight<=2.5):
        return 2.5
    elif(weight>2.5 and weight<=3.0):
        return 3
    elif(weight>3.0 and weight<=3.5):
        return 3.5
    elif(weight>3.5 and weight<=4.0):
        return 4
    elif(weight>4.0 and weight<=4.5):
        return 4.5
    else:
        return 0
```
